## Task-4 Analysis of traffic accident data

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df5= pd.read_csv("RTA Dataset.csv")

df5.head()
df5.describe()
```

```python
df5.isnull().sum()
```

```python
df5.bfill(inplace=True)

df5.isnull().sum()
```

```python
df5['Hour'] = pd.to_datetime(df5['Time'], format='%H:%M:%S').dt.hour
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Hour', data=df5, hue='Hour', palette='viridis', legend=False)
plt.title('Number of Accidents by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Day_of_week', data=df5, hue='Day_of_week',palette='viridis', legend=False)
plt.title('Number of Accidents by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Age_band_of_driver', hue='Age_band_of_driver',data=df5, palette='viridis')
plt.title('Number of Accidents by Age Band of Driver')
plt.xlabel('Age Band of Driver')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Type_of_vehicle', hue='Type_of_vehicle',data=df5, palette='viridis')
plt.title('Number of Accidents by Type of Vehicle')
plt.xlabel('Type of Vehicle')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=65, ha='right') 
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='Cause_of_accident', data=df5, palette='viridis', hue='Cause_of_accident',order=df5['Cause_of_accident'].value_counts().index)
plt.title('Causes of Accidents')
plt.xlabel('Number of Accidents')
plt.ylabel('Cause of Accident')
plt.show()
```

## Analysis-
I.According to above representation major accidents are happened during evening period that is between 4pm to 7pm.  
II.Most accidents are occured during weekdays than weekend.  
III.Ages from 18 to 50 have done more accidents than other age group.  
IV.AUtomobile annd Lorry this type of vehicles have large numbers of accidents.  
V.Causes of accidents are due to No Distancing, Changing lanes, driving carelessly and other.
